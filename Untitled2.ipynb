```python
# 필요한 라이브러리 임포트
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일 읽기
raw_data = pd.read_csv('/content/survey_results_public.csv')  # Colab에서는 보통 '/content/' 경로에 업로드됨

# 필요한 열 추출
target_data = raw_data[['Age', 'Country', 'LanguageHaveWorkedWith', 'LearnCode']]

# 결측치 제거 후 문자열 분리
methods = target_data['LearnCode'].dropna()
methods = methods.str.split(';')

# 분리된 데이터 펼치기
exploded_methods = methods.explode()

# 학습 방법별 개수 계산
size_by_methods = exploded_methods.groupby(exploded_methods).size()

# 상위 10개 항목을 파이 차트로 시각화
top10_methods = size_by_methods.nlargest(10)
plt.figure(figsize=(5, 5))
top10_methods.plot.pie(autopct='%1.0f%%')
plt.ylabel('LearnCode')
plt.title('Top 10 Learning Methods of Developers')
plt.tight_layout()
plt.show()
```
